## Como percorrer um dataset usando loops (for)

Para isso, você pode usar a ideia de um loop, onde ele vai percorrer cada linha do seu dataset. Esse conceito vai ser bastante utilizando, principalmente na plotagem de um gráfico. No exemplo abaixo estamos percorrendo todo dataset (linha por linha), porém, "armazenando" (usando uma lista) apenas a coluna "OFFENSE_CODE"

```python
lista = []
for offense_code in df["OFFENSE_CODE"]:
    lista.append(offense_code)
```

Esse trecho faz algo parecido: 

```python
lista_2 = []
lista_2.append(df["OFFENSE_CODE"])
lista_2
```

## Criando um contador sem usar funções do próprio pandas

Esse código é um exemplo do que você pode fazer sem usar o pandas. É uma função similar ao value_count, porém, estou apenas contando um elemento do meu dataset. No caso, quero contar as vezes que o DISTRICT = D14 aparece no meu dataset. Você pode usar o pandas, mas isso é apenas para ter uma noção de como fazer de outra forma

```python
count_d4 = 0 #Iniciando um contador com 0 (declarando a variavel)
for district in df["DISTRICT"]: #Iniciando o for para percorrer toda a coluna DISTRICT do nosso dataset
    
    # Esse trecho diz que a cada district D4, vamos incrementar o nosso contador
    if district == "D4":
        count_d4 += 1
        
print('Número de elementos D4: ', count_d4)
```

Mesmo principio do contador mostrado, porém, quero contar todas as ocorrencias dos elementos (Esse é a mesma coisa do value count do pandas). A ideia aqui é usar uma lista chave: valor. Esse tipo de estrutura funciona da seguinte forma (como um dicionário)

- {'dia': 'significado da palavra dia',
 'noite': 'significado da palavra noite'}
 
Fazendo uma analogia a um dicionário, eu posso saber o significado da palavra dia apenas procurando essa palavra no dicionário. Em Python, você acessaria dessa forma:

- lista['dia'], onde 'dia' é uma chave para procurar um determinado valor.

```python
count = {}
for district in df["DISTRICT"]:
    if district not in count:
        count[district] = 0
    count[district] += 1
```

## Criando gráficos de Pizza
Para fazer isso com matplotlib é bem simples:

- plt.pie("Aqui vão os seus dados", labels="Nome dos seus dados", colors="Cor de cada um no grafico")


```python

plt.pie([10, 2, 3], labels = ['malware', 'trusted', 'riskware'], 
            colors = ['red', 'blue', 'green'], 
            autopct='%1.1f%%', startangle = 90)

plt.title("Segurança digital")
plt.show()
```

Com isso, você pode mostrar o mesmo gráfico para ver o número de ocorrências que você tem:

```python
count = {}
for district in df["DISTRICT"]:
    if district not in count:
        count[district] = 0
    count[district] += 1
```

```python
#count.values() <- Esse mostra apenas os valores de dentro do dicionario chave: valor
#count.items() <- Esse mostra apenas as categorias

# Note que aqui não definimos COLORS. Isso quer dizer que isso será definido automático
plt.pie(count.values(), labels = count.items(), 
            autopct='%1.1f%%', startangle = 90)

plt.title('Crimes (DISTRICT)')
plt.show()
```

Talvez alguns gráficos não demonstrem bem os dados. Por isso, você precisa tentar modificar o tipo para ver qual representa melhor a relação do seu dataset. Vou mostrar os mesmos dados acima, porém, com um gráfico diferente:

```python
# Aqui eu faço a contagem com o pandas
numbers_of_district = df["DISTRICT"].value_counts()
# Aqui eu chamo o meu grafico de barras
numbers_of_district.plot(kind='bar')

# Titulo do grafico
plt.title(" Números de crimes por distrito (Boston)")

# Texto eixo x
plt.xlabel(" Distritos")

# Texto eixo y
plt.ylabel(" Número de ocorrências (crimes)")
# Aqui eu mostro o meu grafico de barras
plt.show()
```

Pensando no seu dataset, como você pode compara-lo? Tente fazer uma correlação (em um gráfico) de contagem, ou seja, quantas ocorrências "mobile" e quantas mudanças "mobile" você teve em um determinado dia? 

Exemplo de um gráfico de barras:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np



labels = ['02/02/2020', '03/02/2020', '04/02/2020']
numero_ocorrencias_mobile = [20, 50, 1]
numero_mudancas_mobile = [25, 4, 2]

x = np.arange(len(labels))  # Posição dos textos
width = 0.35  # Tamanho da barra

fig, ax = plt.subplots() # Chamando os subgráficos

rects1 = ax.bar(x - width/2, numero_ocorrencias_mobile, width, label='ocorrencias') # Primeira barra
rects2 = ax.bar(x + width/2, numero_mudancas_mobile , width, label='mudanças') # Segunda barra


ax.set_ylabel(' Ocorrências ')
ax.set_title(' Grupo de ocorrências')
ax.set_xticks(x)
ax.set_xticklabels(labels, size = 10)
ax.legend()
```
