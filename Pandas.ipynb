## Condição para filtragem

A sintaxe correta para filtrar elementos do pandas é essa aqui. Nesse caso você não precisa converter para "int". Provavelmente o erro ocorreu por algum erro na base. Tente fazer dessa forma para ver se ele consegue filtrar os elementos: 

Sintaxe correta: DF[DF["NOME_COLUNA"] < CONDIÇAO]]

```python
df[df["REPORTING_AREA"] < "347"]
```

## Filtrando elementos (mais de uma coluna)

Separar elementos "REPORTING_AREA" < "347" E "OFFENSE_CODE_GROUP" == "Towed"

```python
df[(df["REPORTING_AREA"] < "347") & (df["OFFENSE_CODE_GROUP"] == "Towed")]
```

Separar elementos "REPORTING_AREA" < "347" OU "OFFENSE_CODE_GROUP" == "Towed"

```python
df[(df["REPORTING_AREA"] < "347") | (df["OFFENSE_CODE_GROUP"] == "Towed")]
```
